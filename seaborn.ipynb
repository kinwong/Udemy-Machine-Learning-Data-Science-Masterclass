```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('notebooks_final/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=240)
sns.scatterplot(x='salary', y='sales', data=df,hue='level of education', palette='Dark2', size='work experience', alpha=0.4, style='level of education');
# plt.savefig('scatter.png')
```

```python
plt.figure(figsize=(12,4),dpi=240)
sns.rugplot(x = 'salary', data=df, height=0.5);
```

```python
plt.figure(figsize=(20,4),dpi=240)
sns.set_style('darkgrid')
sns.displot(x = 'salary', data=df, bins=20, color='red', kde=True);
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='division', hue='level of education');
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df, x='level of education', y='salary', hue='division', estimator=np.mean, ci='sd')
plt.legend(bbox_to_anchor=(1.05, 1));
```

```python
df = pd.read_csv('notebooks_final/05-Seaborn/StudentsPerformance.csv')
df.head()
```

```python
fig = plt.figure(figsize=(12,6),dpi=240)
sns.boxplot(data=df, x='math score', y='parental level of education', hue='test preparation course', palette='Set2');
plt.legend(bbox_to_anchor=(1.05, 0.5));
```

```python
fig = plt.figure(figsize=(10,7),dpi=240)
sns.violinplot(data=df, x='math score', y='parental level of education', hue='test preparation course', split=True, inner='quartile', bw=0.1);
plt.legend(bbox_to_anchor=(1.25, 0.5));
```

```python
fig = plt.figure(figsize=(10,5),dpi=240)
sns.swarmplot(data=df, x='math score', y='gender', hue='test preparation course', dodge=True);
```

```python
fig = plt.figure(figsize=(10,5),dpi=240)
sns.boxenplot(data=df, x='math score', y='race/ethnicity', hue='test preparation course');
```

```python
fig = plt.figure(figsize=(10,5),dpi=240)
sns.jointplot(data=df, x='math score', y='reading score', kind='hist', alpha=0.8, hue='gender');
```

```python
fig = plt.figure(figsize=(10,5),dpi=240)
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', shade=True, alpha=0.5, hue='test preparation course');
```

```python
fig = plt.figure(figsize=(10,5),dpi=240)
sns.pairplot(data=df, hue='gender', corner=True)
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin', row='lunch', col='test preparation course')
```

```python
g = sns.PairGrid(df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
```
