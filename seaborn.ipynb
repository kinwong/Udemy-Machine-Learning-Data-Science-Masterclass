```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('notebooks_final/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=240)
sns.scatterplot(x='salary', y='sales', data=df,hue='level of education', palette='Dark2', size='work experience', alpha=0.4, style='level of education');
# plt.savefig('scatter.png')
```

```python
plt.figure(figsize=(12,4),dpi=240)
sns.rugplot(x = 'salary', data=df, height=0.5);
```

```python
plt.figure(figsize=(20,4),dpi=240)
sns.set_style('darkgrid')
sns.displot(x = 'salary', data=df, bins=20, color='red', kde=True);
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='division', hue='level of education');
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df, x='level of education', y='salary', hue='division', estimator=np.mean, ci='sd')
plt.legend(bbox_to_anchor=(1.05, 1));
```
