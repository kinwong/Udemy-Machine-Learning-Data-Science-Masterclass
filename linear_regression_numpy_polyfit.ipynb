```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('notebooks_final/08-Linear-Regression-Models/Advertising.csv')
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
df.head()
```

```python
fig = plt.figure(figsize=(10, 8), dpi=200)
sns.regplot(data=df, x='total_spend', y='sales');
```

```python
X = df['total_spend']
y = df['sales']
# help(np.polyfit)
# y = mx + b
# y = B1x + b0
poly = np.polyfit(X, y, deg=1)
poly
```

```python
potential_spend = np.linspace(0, 500, 100)
potential_sales = poly[0] * potential_spend + poly[1]

fig = plt.figure(figsize=(10, 8), dpi=200)
sns.regplot(data=df, x='total_spend', y='sales');
plt.plot(potential_spend, potential_sales, color='red', linewidth=2);
```

```python
spend = 200
sales = poly[0] * spend + poly[1]
sales
```

```python
# y = B3x**3 + B2*x**2 + B1*x + B0
poly = np.polyfit(X, y, deg=3)
poly
```

```python
pot_spend = np.linspace(0, 500, 100)
pred_sales = poly[0] * pot_spend**3 + poly[1] * pot_spend**2 + poly[2] * pot_spend + poly[3]
```

```python
potential_spend = np.linspace(0, 500, 100)
potential_sales = poly[0] * potential_spend + poly[1]

fig = plt.figure(figsize=(10, 8), dpi=200)
sns.regplot(data=df, x='total_spend', y='sales')
plt.plot(pot_spend, pred_sales, color='green', linewidth=2)
```
