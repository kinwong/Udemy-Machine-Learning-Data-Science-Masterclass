```python
a = np.linspace(0, 10, 11)
a
```

```python
b = a ** 4
b
```

```python
x = np.arange(0, 10)
x
```

```python
y = 2*x
y
```

```python
fig = plt.figure(figsize=(2,1.5), dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('Power of 4')
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.plot(a, b, label='a vs b')

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.set_title('Zoomed In')
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_xlim(2, 3)
axes2.set_ylim(0, 100)

axes2.plot(a ,b)

fig.savefig('test.png', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4,3), dpi=200)
axes[0,0].plot(x, y, label='x vs y')
axes[0,0].plot(a, b, label='a vs b')
axes[0,0].legend()

axes[0,1].plot(a, b)
axes[1,0].plot(x, y)
axes[1,1].plot(a, b)
fig.subplots_adjust(wspace=0.5, hspace=0.5)
# plt.tight_layout()
```

```python
x = np.linspace(0,11,10)
x
```

```python
fig = plt.figure(facecolor='lightyellow', dpi=140)
ax = fig.add_axes([0,0,1,1])
line = ax.plot(x,x, color='blue', label='x vs x', lw=4, ls='-.', marker='+', markersize=20, markerfacecolor='orange')
line[0].set_dashes([5,1,10,1])
ax.plot(x, x ** 2, color='#223454', label='x vs x^2',linewidth=0.5, linestyle='--', marker='s', mfc='green')
ax.legend()
```
