```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title('x vs y')
plt.xlabel('x values')
plt.ylabel('y values')
plt.xlim(-5, 5)
plt.ylim(0, 10)
plt.show()
```

```python
# help(plt.savefig)
plt.savefig('test.png');
```

```python
fig = plt.figure()
axe = fig.add_axes([0.5, 0.5, 2, 1])
axe.plot(x,y)
```
